**EVALUACIÓN 1 DE CIENCIA DE DATOS ESPACIALES


```python
#Estos son las 3 fuentes de datos de países 

electricityLink='https://www.worldometers.info/energy/'
populationLink= 'https://www.cia.gov/the-world-factbook/field/population/country-comparison/'
co2Link='https://www.worldometers.info/co2-emissions/co2-emissions-by-country/'

import pandas as pd
co2List =pd.read_html(co2Link, header=0,flavor='bs4')
electricityList =pd.read_html(electricityLink, header=0,flavor='bs4')
populationList =pd.read_html(populationLink, header=0,flavor='bs4')

co2 = co2List[0]
electricity = electricityList[0]
population = populationList[0]
```

```python
#merging

co2.shape, population.shape, electricity.shape
```

```python
#nombre de las clumnas de los dataframes

co2.columns, population.columns, electricity.columns
```

```python
#La cantidad de coincidencias entre co2 y population que tendra el merge

co2.merge(population,how='inner',left_on='Country',right_on='Country').shape
```

```python
co2.merge(electricity,how='inner',left_on='Country',right_on='Country').shape
```

```python
electricity.merge(population,how='inner',left_on='Country',right_on='Country').shape
```

```python
#Realizamos el merge

combinacion_1=co2.merge(population,left_on='Country',right_on='Country')
combinacion_1
```

```python
#vemos las concidencias entre combinacion_1 y electricity

electricity.merge(combinacion_1,how='inner',left_on='Country',right_on='Country').shape
```

```python
#realizamos el merge entre la primera combinación y el 3er dataframe

combinacion_2=electricity.merge(combinacion_1,left_on='Country',right_on='Country')
combinacion_2
```

```python
#Ahora realizaremos el merge con los datos del profesor 
import os
tabla_profesorLink='https://github.com/juandiegopozo-10/evaluacion1datos/blob/main/country_isos.xlsx'
import pandas as pd 
Tabla_1 = pd.read_excel(os.path.join('country_isos.xlsx'))
Tabla_1
```

```python
# ahora convertiremos todos los paises del primer merge (combinacion_2) ha mayúscula para que así haya coincidencias

combinacion_2['Country'] = combinacion_2['Country'].str.upper()
combinacion_2
```

```python
#ahora sí realizamos el merge con la tabla proporcionada por el profesor 

Tabla_1.merge(combinacion_2,how='inner',left_on='Country',right_on='Country').shape
```

```python
combinacion_final = combinacion_2.merge(Tabla_1,left_on='Country',right_on='Country')

#Este seriasería el resultado final 
combinacion_final
```

```python
#por último limpiamos un poco la data
combinacion_final.columns = combinacion_final.columns.str.replace(' ', '_')

combinacion_final
```

```python
combinacion_final = combinacion_final.sort_values(by='Country')
combinacion_final
```
